#### EJERCICIO 5
Tenemos datos de hombres y mujeres por edades, obten el diagrama de barras apiladas.

Edades hombres: 29, 29, 24, 17, 13,  7, 11, 10, 12, 13,  9,  5, 23, 14, 16
Edades mujeres: 5,  11, 23,  6, 25, 16, 10, 27,  8, 27, 15, 29, 21, 10, 28

```python
unique, counts = np.unique([29, 29, 24, 17, 13,  7, 11, 10, 12, 13,  9,  5, 23, 14, 16], return_counts=True)
unique2, counts2 = np.unique([ 5,  11, 23,  6, 25, 16, 10, 27,  8, 27, 15, 29, 21, 10, 28], return_counts=True)
```

```python
x = unique
y = counts

x2 = unique2
y2 = counts2

fig, ax = plt.subplots()



plt.bar(x,y, label = 'grupo A')
plt.bar(x2,y2, bottom= y,label = 'grupo B')
ax.legend()
plt.show()
```

**CORRECCI√ìN**

```python
import pandas as pd
edad_M = pd.Series([29, 29, 24, 17, 13,  7, 11, 10, 12, 13,  9,  5, 23, 14, 16])
edad_F = pd.Series([5,  11, 23,  6, 25, 16, 10, 27,  8, 27, 15, 29, 21, 10, 28])

df_freq_edades = pd.concat([edad_M.value_counts(), edad_F.value_counts()], axis=1, )
df_freq_edades = df_freq_edades.fillna(0)
df_freq_edades.columns = ['M','F']

plt.figure()
plt.bar(df_freq_edades.index,
        df_freq_edades.F,
        label = 'Mujeres')
plt.bar(df_freq_edades.index,
        df_freq_edades.M,
        bottom = df_freq_edades.F,
        label = 'Hombres')
plt.show()
```

#### EJERCICIO 10

Utilizando de nuevo el conjunto de datos **df_iris**, realiza un gr√°fico de sectores para cada variable, para los datos pertenecientes a la especie = setosa

```python
variables = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']

for values in variables:

    data = df_iris[df_iris['species']=='setosa'][values]
    plt.figure()

    plt.pie(data, labels=data)
    plt.show()
```
