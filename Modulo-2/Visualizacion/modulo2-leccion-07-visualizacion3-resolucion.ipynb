```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np # Ya que usaremos algunas de sus funciones ;)
```

```python
df_iris = sns.load_dataset("iris")
df_penguin = sns.load_dataset("penguins")
```

**EJERCICIO 1**

Crea una gráfoca lineal haciendo uso del dataset iris, de forma que representes la longitud y anchura de los pétalos para las diferentes especies de flores del dataset.


Opcional: juega con algunas de sus opciones de personalización mirando la documentación oficial.

```python
plt.figure(figsize =(8,4))
fig = sns.lineplot(x='petal_length' , y='petal_width' , data = df_iris, ci=None, hue ='species')
fig.set(xlabel='Longitud del pétalo', ylabel = 'Anchura del pétalo')
fig.set_title('Gráfica lineal')
plt.show()
```

### EJERCICIO 3

Haciendo uso de nuevo del dataset **df_iris**, obten los histogramas para las variables petal_width, consideredando el número de bins que consideres adecuados. Incluyendo la informacion correcta en los ejes y las diferentes especies de flores.

#### EJERCICIO 5
Haciendo uso del dataset **df_iris** obten una gráfica de enjambre de la longitud del sépalo por cada una de las especies disponibles.

```python
plt.figure(figsize =(8,4))
fig = sns.swarmplot(x='species', y = 'sepal_length',  data = df_iris)
fig.set(xlabel='Especies', ylabel= 'Longitud del sépalo por especie' )
fig.set_title('Diagrama de enjambre de la longitud del sépalo por especie de flor')
plt.show()
```

#### EJERCICIO 7
Haciendo uso de nuevo del dataset **df_iris**, obten el gráfico de catplot para las variables 'sepal length (cm)'.

```python
plt.figure(figsize =(8,4))
fig = sns.catplot(x='species', y ='sepal_length', data = df_iris, hue = 'species')
fig.set(xlabel='Especie', ylabel= 'Longitud del  sépalo en cm' )
fig.fig.suptitle('Catplot de varible categórica longitud del sépalo por especie')
plt.show()

plt.figure(figsize =(8,4))
fig = sns.catplot(x='species', y ='sepal_width', data = df_iris, hue = 'species', kind = 'box')
fig.set(xlabel='Especie', ylabel= 'Longitud del  sépalo en cm' )
fig.fig.suptitle('Catplot de varible categórica longitud del sépalo por especie. con cajas')
plt.show()
```
