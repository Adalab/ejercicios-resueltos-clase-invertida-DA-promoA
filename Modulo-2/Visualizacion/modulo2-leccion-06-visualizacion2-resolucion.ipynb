Representa sobre la misma figura las 3 gráficas para los diagramas de dispersion de las diferentes especies de flores. De las siguentes 2 formas:
 - Las 3 gráficas en vertical
 - Las 3 gráficas en horizontal

```python

# VERTICAL
mapa={'setosa':0,'versicolor':1,'virginica':2}
df_iris['species_num'] = df_iris['species'].replace(mapa)


df_iris_setosa = df_iris[df_iris['species'] == 'setosa']
df_iris_virginica = df_iris[df_iris['species'] == 'virginica']
df_iris_versicolor = df_iris[df_iris['species'] == 'versicolor']


fig,ax = plt.subplots(3,1, constrained_layout=True)

ax[0].scatter(df_iris_setosa['sepal length (cm)'],df_iris_setosa['sepal width (cm)'], c = 'violet', label = 'setosa', marker='d', s= 8)
ax[1].scatter(df_iris_virginica['sepal length (cm)'],df_iris_virginica['sepal width (cm)'], c = 'yellow',label = 'virginica', marker = '.', s= 10)
ax[2].scatter(df_iris_versicolor['sepal length (cm)'],df_iris_versicolor['sepal width (cm)'], c = 'blue', label = 'versicolor', marker = '^', s= 10)
plt.xlim([4,9])
plt.ylim([2,4.5])

ax[0].grid(color ='black', linestyle = '--', linewidth = 0.5)
ax[1].grid(color ='black', linestyle = '--', linewidth = 0.5)
ax[2].grid(color ='black', linestyle = '--', linewidth = 0.5)

ax[0].legend()
ax[1].legend()
ax[2].legend()


plt.suptitle('Gráfica de dispersion longitud frente a anchura del sépalo, para todas las especies.')
fig.text(0.5,0.04, "Longitud del sépalo (cm)", ha="center", va="center")
fig.text(-0.05,0.5, "Anchura del sépalo (cm)", ha="center", va="center", rotation=90)
plt.show()
```

```python
# HORIZONTAL
mapa={'setosa':0,'versicolor':1,'virginica':2}
df_iris['species_num'] = df_iris['species'].replace(mapa)


df_iris_setosa = df_iris[df_iris['species'] == 'setosa']
df_iris_virginica = df_iris[df_iris['species'] == 'virginica']
df_iris_versicolor = df_iris[df_iris['species'] == 'versicolor']


fig,ax = plt.subplots(1,3, constrained_layout=True)

ax[0].scatter(df_iris_setosa['sepal length (cm)'],df_iris_setosa['sepal width (cm)'], c = 'violet', label = 'setosa', marker='d', s= 8)
ax[1].scatter(df_iris_virginica['sepal length (cm)'],df_iris_virginica['sepal width (cm)'], c = 'yellow',label = 'virginica', marker = '.', s= 10)
ax[2].scatter(df_iris_versicolor['sepal length (cm)'],df_iris_versicolor['sepal width (cm)'], c = 'blue', label = 'versicolor', marker = '^', s= 10)
plt.xlim([4,9])
plt.ylim([2,4.5])

ax[0].grid(color ='black', linestyle = '--', linewidth = 0.5)
ax[1].grid(color ='black', linestyle = '--', linewidth = 0.5)
ax[2].grid(color ='black', linestyle = '--', linewidth = 0.5)

ax[0].legend()
ax[1].legend()
ax[2].legend()


plt.suptitle('Gráfica de dispersion longitud frente a anchura del sépalo, para todas las especies.')
fig.text(0.5,-0.04, "Longitud del sépalo (cm)", ha="center", va="center")
fig.text(-0.05,0.5, "Anchura del sépalo (cm)", ha="center", va="center", rotation=90)
plt.show()
```
