```python
import pandas as pd
```

```python
df = pd.read_csv("datos/netflix_titles.csv")
df.head()
```

```python
# para sacar el números de filas y columnas

df.shape
```

```python
# para sacar los tipos de datos

df.dtypes
```

```python
# para sacar los nulos en porcentaje

df.isnull().sum() * 100 / df.shape[0]
```

```python
# los principales estadísticos.En este caso son todas numéricas por lo que no tendremos que especificar nada más

df.describe()
```

```python
# primero sacamos las variables categóricas

categoricas = df.select_dtypes(include = "object")
categoricas.head(2)
```

```python
# valores únicos

for col in categoricas.columns:
    print("------------------------------------------------------------", "\n")
    print(f"En la columna {col} tenemos los siguientes valores unicos")
    print(pd.DataFrame(df[col].value_counts()))
    print("\n", "------------------------------------------------------------", "\n")
```

```python
for col in categoricas.columns:
    print("------------------------------------------------------------", "\n")
    print(f"En la columna {col} tenemos los siguientes valores unicos")
    print(pd.DataFrame(df[col].value_counts()))
    print("\n", "------------------------------------------------------------", "\n")
```
