1️⃣ Crea un dataframe con aquellos clientes que el `sales_channel`sea a través de un *call centrer* y vivan en el entorno rural. 

2️⃣ Usa el método `str.contains` para extraer aquellos vehiculos que tengan dos puertas. 

3️⃣ Usa el método `np.select` para filtrar por los clientes que responden o que el número de polizas sea mayor a 5 y crear una columna nueva. Aquellos que cumplan la condición les asignamos el valor de "Top" y a los que no "Down". 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("datos/Marketing-Customer-Analysis.csv")
df.head(2)
```

```python
# cambiamos el nombre de las columnas

nuevas_columnas = {col: col.lower().replace(" ", "_") for col in df.columns}
nuevas_columnas
```

```python
df.rename(columns = nuevas_columnas, inplace = True)
```

```python
# 1️⃣ filtro por sales_channel y location_code

df_loc_sales = df[(df["location_code"] == "Rural") | (df["sales_channel"] == "Call Center")]
df_loc_sales.head(2)
```

```python
# 2️⃣ Vehiculos con dos puertas

patron = "Two"

dos_puertas = df[df["vehicle_class"].str.contains(patron, regex = True)]
dos_puertas.head(2)
```

```python
# 3️⃣ clientes que contestan con más de 5 polizas

condiciones = ((df["response"] ==  "No") | (df["number_of_policies"] > 5)), ((df["response"] !=  "No") | (df["number_of_policies"] < 5)) 

opciones = ["Top", "Down"]
```

```python
df["nueva"] = np.select(condiciones, opciones)
```
