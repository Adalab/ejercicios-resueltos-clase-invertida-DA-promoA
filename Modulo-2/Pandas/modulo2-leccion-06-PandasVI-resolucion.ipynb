```python
import pandas as pd
```

```python
# 1️⃣ Cargar el dataset

df = pd.read_csv("datos/titanic.csv")
df.head()
```

```python
# 2️⃣ ¿Cuánta gente sobrevivió de media en función de la clase? 

df.groupby(["pclass"])["survived"].mean().reset_index()
```

```python
# 3️⃣ ¿Hubo diferencia entre sexos y clases en la media de personas que sobrevivieron? 

df.groupby(["pclass", 'sex'])["survived"].mean().reset_index()
```

Independientemente de la clase, se observa una tendencia a que sobrevivían más mujeres que hombres

```python
# 4️⃣ Calcula la media, la mediana y mediana de la edad en funcion de la clase. 

df.groupby("pclass")["age"].agg(['mean', 'std', 'median'])
```
