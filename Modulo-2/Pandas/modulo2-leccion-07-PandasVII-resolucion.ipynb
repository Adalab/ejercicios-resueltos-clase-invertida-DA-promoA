```python
import pandas as pd
```

```python
# 1️⃣ Carga el dataset
df = pd.read_csv("datos/titanic.csv")
df.head()
```

```python
df2 = df.copy()
```

```python
# 2️⃣ Crea una función que usando la columna `name` nos cree una columna nueva que nos diga si el pasajero era Miss, Master, Mrs, etc.

# hay un nulo, por eso hay que incluir el try, except. 
def sacar_posicion(x):
    
    try:
        return x.split(".")[0].split(",")[1]
    except: 
        return x
```

```python
df2["posicion"] = df2["name"].apply(sacar_posicion)
```

```python
df2.head(2)
```

```python
# 3️⃣ La columna `survived` esta con valores numéricos. Convierte los 1 a Si y los 0 a No. 
def convertir(x):
    if x == 1:
        return "Si"
    else:
        return "No"
```

```python
df2["vivo/muerto"] = df2["survived"].apply(convertir)
```

```python
df2.head(2)
```

```python
# 4️⃣  Guarda en un excel el dataframe resultante en una hoja y en otra el orginal. 
from pandas import ExcelWriter

w = ExcelWriter("datos/titanic2.xlsx") 


df2.to_excel(w, sheet_name="limpio") 
df.to_excel(w, sheet_name="sucio") 
    
```
