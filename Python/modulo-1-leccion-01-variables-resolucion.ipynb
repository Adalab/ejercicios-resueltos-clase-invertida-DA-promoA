## Ejercicios: operaciones algebraicas

- Operaciones básicas

  14. Calcula la suma de `gatos` y `pizzas`.

  16. Multiplica las cantidad de gatos por 2.
  17. Haz que la cantidad de `piezas` sea 8 veces la cantidad de `pizzas`.
  18. Haz que la cantidad de `piezas` se aumente con 3.

  20. ¿Cuanta superficie hay por gato?

---



```python
print('24:', gatos > pizzas)
print('25:', gatos < superficie)
print('26:', 2*pizzas != gatos)
print('27:', 2*pizzas == gatos)

print('28:', piezas >= pizzas and superficie != 0)
print('29:', pizzas <= piezas and piezas < superficie)
print('30:', pizzas > gatos or piezas % gatos == 0)

# Bonus
print('31:', pizzas <= piezas < superficie)
print('32:', bool(superficie))
```
