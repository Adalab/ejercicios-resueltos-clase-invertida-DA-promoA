## Contar matches regex con una función

```python
def contar_patrones(patron,string):
    # patron = patron_e
    # patron = 'e'
    # string = libro
    patrones = re.findall(patron,string)

    return len(patrones)
```

```python
patron_e = 'e'
cantidad_e = contar_patrones(patron_e,libro)
cantidad_a = contar_patrones('a',libro)
```

```python
print(cantidad_e)
print(cantidad_a)
```

```python
los_quijotes = contar_patrones('\WQ[Uu][Ii][Jj][Oo][Tt][Ee]\W',libro)
print(los_quijotes)
```

```python
los_quijanos = contar_patrones('\WQ[Uu][Ii][Jj][Aa][Nn][Oo]\W',libro)
print(los_quijanos)
```

## Counter

```python
from collections import Counter
import re

# ¡Cuidado!

print('original:',libro[0:45])

patron = '\w+'
lista_palabras = re.findall(patron,libro)

print('lista_p:',lista_palabras[0:5])
```

```python

contenedor = lista_palabras
top_n = 25
c = Counter(contenedor).most_common(top_n)

print(c)
```
