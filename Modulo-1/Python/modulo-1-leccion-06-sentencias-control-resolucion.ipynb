  21. Comprueba si `var1` es más pequeño que 3, si no es así, mira si es màs grande que 15. En el primer caso, imprime `Qué pequeño es.`, en el segundo caso imprime `¡Tan grande!`. En los demás casos imprime `Así me gusta.`.

  23. Cambia el código que hiciste en la 22. para que el rango entre `3` y `var3` imprimiría la frase `Así me gusta.`. Las frases para los valores más pequeños y más grandes se aplican igual que en la 22. ¿Ha cambiado uno (o ambos) de los rangos de valores que corresponden a esas frases?

---



```python
var1 = 8
var2 = 2
var3 = 8

if var1 < 3:
  print('Qué pequeño es.')
elif var1 > 15:
  print('¡Tan grande!')
else:
  print('Así me gusta.')

if var1 > 3:
  print('¡Tan grande!')    # valores más grande que 3, (3,inf]
elif var1 <= var3:
  print('Qué pequeño es.') # valores más pequeño que 3, [-inf,3] si var3 > 3, [-inf, var3] si var3 < 3
else:
  print('Nunca se llegaría aquí, porque var3 > 3.') # si var3 fuera < 3 aquí entrarían los valores entremedio

if var1 < 3:
  print('Qué pequeño es.') # valores más pequeño que 3, la frase ya no apunta a valores entre 3 y var3
elif var1 > var3:
  print('¡Tan grande!')    # valores más grandes que var3, la frase ya no apunta a valores entre 3 y var3
else:
  print('Qué pequeño es.') # valores entremedio
```

## Ejercicios `while:` `break`

- Bucles `while`

    32. Inicia `ahorros` con un valor de `5`. Mientras los ahorros sean igual a o más pequeño que `20` imprime la cantidad de ahorros, y añádele `5` más.
    33. Si cambias el orden de imprimir y añadir `5`, ¿qué resultado te esperas? Compruébalo.

- Bucles `while` con `break`

    34. Si los ahorros valen `12` o más, ya no me subirán más. Añade esta información con un break.

---

```python
# 32.
print('--- 32. ---')
ahorros = 5
while ahorros <= 20:
    print(ahorros)
    ahorros += 5


# 33.
print('--- 33. ---')
ahorros = 5
while ahorros <= 20:
    ahorros += 5
    print(ahorros)

# 34.
print('--- 34. ---')
ahorros = 5
while ahorros <= 20:
    print(ahorros)
    ahorros += 5
    if ahorros >= 12:
        break
    
```

## Ejercicios `try:` `except:` `else:` `finally:`

- `try:`, `except:`

    35. Defina `doce = '12'` y `seis = 6`. Intenta sumirlos. En el caso de que te salga un error imprime `'¡Un error!'`
    36. En el ejercicio 35. haz que imprime qué error ha tenido.
    37. Cambia tu solución de 36. para que devuelva el *tipo*  de error que ha sido.
    38. Especifíca un except con el tipo del ejercicio 37. *Dentro* de ese except, cambia `doce + seis` por `int(doce) + int(seis)` en un `print()` y mira qué pasa.

- `else:`, `finally:`

    39. En el caso de que 35. te salga sin errores, querrás imprimir `'Todo bien por aquí.'` Cambia la variable `doce` para llegar a imprimirlo.
    40. Amplia 39. con 'Se acabó el try.' que te sale siempre al acabar el try.

---

```python
doce = '12'
seis = 6

# 35.
try: 
    doce + seis
except:
    print('35. ¡Un error!') 

# 36.
try: 
    doce + seis
except Exception as e:
    print('36. ¡Un error!', e) 

# 37.
try: 
    doce + seis
except Exception as e:
    print('37.', type(e)) 

# 38.
try: 
    doce + seis
except TypeError:
    print('38.', int(doce)+int(seis) )

# 39.
doce = 12
try: 
    doce + seis
except:
    print('39. ¡Un error!') 
else:
    print('39. Todo bien por aquí.')

# 40.
try: 
    doce + seis
except:
    print('40. ¡Un error!') 
else:
    print('40. Todo bien por aquí.')
finally:
    print('    Se acabó el try.')
```

## Ejercicios `continue`, `else`

- Saltar iteraciones y terminar un `while`

    41. Inicia `ahorros` a `5`. Mientras `ahorros` sea menos que `50` lo quiero subir con `10` y imprimir el valor. Si el valor es `25` no lo imprimes.
    42. Amplia 41. para que te salga `'Tengo más que 50 en ahorros, dejo de aumentarlo.'`

---

```python
# 41.
print('--- 41. ---')
ahorros = 5
while ahorros < 50:
    ahorros += 10
    if ahorros == 25:
        continue
    print(ahorros)

# 42.
print('--- 42. ---')
ahorros = 5
while ahorros < 50:
    ahorros += 10
    if ahorros == 25:
        continue
    print(ahorros)
else:
    print('Tengo más que 50 en ahorros, dejo de aumentarlo.')
```
