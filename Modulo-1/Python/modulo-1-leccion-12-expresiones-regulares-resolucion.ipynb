## Ejercicios (más) técnicas de strings

- Definir el string con que trabajamos.

    1. Define `str` como `El suelo está enladrillado, ¿quién lo desenladrillará? El desenladrillador que lo desenladrille, buen desenladrillador será.`

- Repaso técnicas de strings.

    2. `str2`: Manipula El String Para Que Las Palabras Empiecen Con Una Mayúscula Todas.
    3. `str3`: Manipula `str2` para que sALGA aSÍ.

- Más técnicas de strings.

    4. Imprime la lista de palabras de `str`. Guárdala como `por_palabras`.
    5. Manipula `str` para que el `'boli'` sea `'enbolsillado'`. Cambia todos los verbos de ese modo.
    6. Junta `por_palabras` para así recuperar el string original.
    7. ¿En qué índice está la palabra `'buen'`?

- Formatear.

    8. Define un string (`str_huecos`) con huecos **vacíos** para `'unidad'` y `5`, para así obtener: "La unidad te sale a 5 euros. Los 5 euros son incl. IVA." ¿Cuántos huecos necesitas?
    9. Cambia el string (`str_huecos`) a huecos **enumerados** para `'unidad'` y `5`, para así obtener: "La unidad te sale a 5 euros. Los 5 euros son incl. IVA." ¿Cuántos huecos necesitas?
    10. Imprime `str_huecos` para manzanas de 3 euros.
    11. ¿Qué tienes que cambiar para que los euros salen como `3.00`? Hazlo.
    12. Define `str_claves` asignando las claves `producto` y `precio` a la frase de 9.

- Bonus

    15. ¿Puedes conseguir 5. en una sola línea, dentro del mismo print?

```python
# 1.
str = "El suelo está enladrillado, ¿quién lo desenladrillará? El desenladrillador que lo desenladrille, buen desenladrillador será."

# 2.
str2 = str.title()
print('2.', str2)

# 3.
str3 = str2.swapcase()
print('3.', str3)

# 4.
por_palabras = str.split()
print('4.', por_palabras)

# 5.
boli = str.replace('suelo','boli')
print('5.', boli.replace('ladrill','bolsill'))

# 6.
print('6.', ' '.join(por_palabras))

# 7.
print('7.', str.find('buen'))

# 8. 
str_huecos = "La {} te sale a {} euros. Los {} euros son incl. IVA."
print('8.', str_huecos.format('unidad',5,5))

# 9. 
str_huecos = "La {0} te sale a {1} euros. Los {1} euros son incl. IVA."
print('9.', str_huecos.format('unidad',5))

# 10. 
print('10.', str_huecos.format('manzana',3))

# 11. 
str_huecos = "La {0} te sale a {1:.2f} euros. Los {1:.2f} euros son incl. IVA."
print('11.', str_huecos.format('manzana',3))

# 12. 
str_claves = "La {producto} te sale a {precio} euros. Los {precio} euros son incl. IVA."
print('12.', str_claves.format(producto='unidad',precio=5))

# 13. 
str_claves = "La {producto} te sale a {precio:.2f} euros. Los {precio:.2f} euros son incl. IVA."
print('13.', str_claves.format(producto='aspiradora',precio=150))

# 14. 
str_claves = "La {producto} te sale a {precio:d} euros. Los {precio:d} euros son incl. IVA."
print('13.', str_claves.format(producto='aspiradora',precio=150))

# Bonus
# 15.
print('15.', str.replace('suelo','boli').replace('ladrill','bolsill'))
```

## Ejercicios funciones de `re`

- Encontrar verbos sobre ladrillos.

    16. Importa el módulo necesario.
    17. Imprime el `str` tal cual. ¿Qué formas de (des)enladrillar ves? ¿Qué tienen en común?
    18. Busca todas las instantes de esos verbos con una expresión regular y una función del módulo `re`.

- Match objects.

    19. ¿De qué índice a qué indice llega la palabra `suelo` en `str`?
    20. ¿De qué índice a qué índice llegan las formas de (des)enladrillar? Usa un bucle.
    21. Saca el índice de todos las `'e'` que haya en la frase, como `ix_e`.
    22. Saca el índice de todos las `'i'` que haya en la frase, como `ix_i1`.
    23. ¿Cuáles de los `'i'` se encuentran dentro de una forma de (des)enladrillar? Saca sus índices como `ix_i2`.

- Sustituir.
    24. Sustituye las formas de (des)enladr**i**llar por desenladr**o**llar usando `re`.
    25. ¿Se te occurre otra manera de conseguir lo mismo sin usar el módulo `re`?

- Trabajar con los resultados.
    
    26. De los ejercicios 22. y 23. tienes dos listas. Conviértelas en otro tipo de contenedor en que te sea fácil encontrar la diferencia entre estos dos listas. ¿Qué elementos hay en la una que no estén en la otra? 
    27. El resultado de 26. ¿qué significa? Explica qué es a un interlocutor que no haya hecho estos ejercicios.

- Bonus.

    28. Resuelve el ejercicio 20. en una sola línea, dentro del mismo print.
---

```python
# 16.
import re

# 17.
print('17.', str)
print('   ', 'enladrillado,', 'desenladrillará,', 'desenladrillador,', 'desenladrille,', 'desenladrillador')
print('   ', 'enladrill')

# 18.
print('18.', re.findall('\w*enladrill\w*',str))

# 19.
print('19.',re.search('\w*enladrill\w*',str).span())

# 20.
print('20.')
for ladr in re.finditer('\w*enladrill\w*',str):
    print(ladr.span())

# 21.
match_e = re.finditer('e',str)
ix_e = []
for match in match_e:
    ix_e.append(match.start())
print('21.', ix_e)

# 22.
match_i1 = re.finditer('i',str)
ix_i1 = []
for match in match_i1:
    ix_i1.append(match.start())
print('22.', ix_i1)

# 23.
match_i2 = re.finditer('enladr(i)ll',str)
ix_i2 = []
for match in match_i2:
    ix_i2.append(match.start() + 'enladrill'.index('i'))
print('23.', ix_i2)

# 24. - 25.
ladroll1 = re.sub('enladrill','enladroll',str)
ladroll2 = str.replace('enladrill','enladroll')
print('24.', ladroll1)
print('25.', ladroll2)

# 26.
print('26.',set(ix_i1)-set(ix_i2))

# 27.
print('27.','En el índice',set(ix_i1)-set(ix_i2),'está la única "i" que NO se encuentra en una forma de (des)enladrillar).')

# Bonus
ladrill = re.finditer('\w*enladrill\w*',str)
print('28.', [ladr.span() for ladr in ladrill])
```
